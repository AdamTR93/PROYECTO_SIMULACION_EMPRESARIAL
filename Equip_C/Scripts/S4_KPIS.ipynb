```python
import pandas as pd
```

## DATAFRAME

```python
df_BANK_marketing = pd.read_csv("S4_df_BANK_marketing.csv")
df_BANK_marketing
```

## 1) ConversiÃ³n DepÃ³sito

```python
clientes_contactados = len(df_BANK_marketing)
clientes_suscritos = df_BANK_marketing['deposit'].sum()
porcentaje_conversion = (clientes_suscritos / clientes_contactados) * 100
print(f'Porcentaje de ConversiÃ³n a DepÃ³sito: {porcentaje_conversion.round(2)}%')
```

## 2) Promedio de la duraciÃ³n de las llamadas de los suscriptores

```python
# Filtrado de los suscriptores
suscriptores = df_BANK_marketing[df_BANK_marketing['deposit'] == 1]


promedio_duracion_llamadas = suscriptores['duration'].mean()
print(f'Promedio de DuraciÃ³n de Llamadas de Suscriptores: {promedio_duracion_llamadas.round(2)} segundos')
```

## 3) Porcentaje de llamadas a telÃ©fono fijo o a mÃ³vil

```python
llamadas_telefono_movil = df_BANK_marketing[df_BANK_marketing['contact'].isin(['telephone', 'cellular'])].shape[0]
total_llamadas = df_BANK_marketing.shape[0]
porcentaje_llamadas = (llamadas_telefono_movil / total_llamadas) * 100
print(f'Porcentaje de Llamadas a TelÃ©fono o MÃ³vil: {porcentaje_llamadas:.2f}%')
```

## 4) Promedio de contactos previos de suscriptores

```python
num_suscriptores = suscriptores.shape[0] #usamos suscriptores del segundo KPI
suma_contactos_previos = suscriptores['previous'].sum()
promedio_contactos_previos = suma_contactos_previos / num_suscriptores
print(f'Promedio de Contactos Previos de Suscriptores: {promedio_contactos_previos.round(2)}')
```

## 5) Mes con mayor tasa de conversiÃ³n

```python
# Agrupar por mes y calcular el porcentaje de conversiÃ³n por mes
conversion_por_mes = df_BANK_marketing.groupby('month')['deposit'].mean() * 100
mes_mayor_conversion = conversion_por_mes.idxmax()
mayor_porcentaje_conversion = conversion_por_mes.max()
print(f'Mes con Mayor Tasa de ConversiÃ³n: {mes_mayor_conversion} ({mayor_porcentaje_conversion.round(2)}%)')
```
