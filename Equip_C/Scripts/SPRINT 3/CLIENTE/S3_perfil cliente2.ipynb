### VARIABLE TRABAJO

```python
# Variable numérica y variable binaria
# Example of the Student's t-test
from scipy.stats import ttest_ind
data1 = cliente['housing']
data2 = cliente['age']
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')

## por el resultado no hay una influencia de la edad en la contratación de hipóteca
```

### VARIABLE JOB

```python
tabla_job_housing = pd.crosstab(cliente['housing'], cliente['job'])
```

```python
tabla_job_housing
```

```python
# Variable categórica y variable binaria
# Example of the Chi-Squared Test
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(tabla_job_housing)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')

## Según el resultado del test, existe una relación entre los diferentes trabajos y que tengan o no hipóteca
```

### VARIABLE MARITAL

```python
tabla_marital_housing = pd.crosstab(cliente['housing'], cliente['marital'])
tabla_marital_housing
```

```python
# Variable categórica y variable binaria
# Example of the Chi-Squared Test
from scipy.stats import chi2_contingency
stat, p, dof, expected = chi2_contingency(tabla_marital_housing)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
	
## Según el resultado del test, existe una relación entre los diferentes estados civiles y que tengan o no hipóteca
```

### VARIABLE BALANCE

```python
# Variable numérica y variable binaria
# Example of the Student's t-test
from scipy.stats import ttest_ind
data1 = cliente['housing']
data2 = cliente['balance']
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')

## por el resultado no hay una influencia del balance en la contratación de hipóteca
```

### ANALISIS MULTIVARIANTE

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Convertir variable binaria a proporciones por grupo
cliente['housing_mean'] = cliente.groupby(['job', 'marital', 'education'])['housing'].transform('mean')

# Modelo ANOVA
modelo = ols('housing_mean ~ C(job) + C(marital) + C(education)', data=cliente).fit()
anova_results = anova_lm(modelo)

print(anova_results)
```

## LOAN

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Convertir variable binaria a proporciones por grupo
cliente['loan_mean'] = cliente.groupby(['job', 'marital', 'education'])['loan'].transform('mean')

# Modelo ANOVA
modelo = ols('loan_mean ~ C(job) + C(marital) + C(education)', data=cliente).fit()
anova_results = anova_lm(modelo)

print(anova_results);
```

CONCLUSIONES:
1) Todas las variables categóricas (job, marital, education) tienen un efecto significativo sobre la variable dependiente (loan), dado que los p-valores son <0.05
2) La variable job tiene el mayor impacto en la variabilidad explicada (𝐹=2606.75 y suma de cuadrados más alta: 
33.57).
3) Residual mean_sq = 0.001171. La variabilidad residual es pequeña, lo que indica que el modelo ANOVA ajusta bien los datos.

Teoria:
df: número de categórias
sum_sq (Suma de cuadrados):Representa la variabilidad explicada por la variable correspondiente.
F (Estadístico F): Valores altos de 𝐹 sugieren que la variable tiene un efecto significativo.
PR(>F) (p-valor): Evalúa la significancia estadística del efecto de la variable. Si el p-valor es menor a 
0.05 se considera que la variable tiene un impacto significativo.

### Modelo de regresión logística para variables binarias (housing, loan, deposit)

```python
import statsmodels.api as sm
import patsy
```

### HOUSING

```python
# Definir la fórmula para el modelo de regresión logística
formula = 'housing ~ C(job) + C(marital) + C(education) + is_in_default + balance'

# Preparar las variables dependientes (y) e independientes (X)
y, X = patsy.dmatrices(formula, cliente, return_type='dataframe')

# Ajustar el modelo logístico
modelo = sm.Logit(y, X).fit()

# Resumen del modelo
print(modelo.summary())
```
